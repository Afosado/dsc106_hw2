```python
import pandas as pd
import numpy as np
```

```python
daily = pd.read_csv('daily_sales.csv')
monthly = pd.read_csv('monthly_sales.csv')
```

```python
daily.head()
```

```python
day_copy = daily.copy().iloc[:28].rename(columns={'Unnamed: 0': 'd'})
day_copy.head()
```

```python
cols = list(day_copy.columns[2:])
cols
```

```python
grouped_day = day_copy.groupby('d')[cols].apply(np.sum).reset_index()
grouped_day
```

```python
grouped_day['d'] = pd.Categorical(grouped_day['d'], ['Monday', 'Tuesday', 'Wednesday', 'Thursday',
                                                     'Friday', 'Saturday', 'Sunday'])
grouped_day = grouped_day.sort_values('d')
grouped_day
```

```python
grouped_day.to_csv('grouped_day.csv')
```

```python
peak_days = pd.DataFrame(grouped_day.set_index('d').apply(np.argmax),
                         columns=['peak_days'])
peak_days['low_days'] = grouped_day.set_index('d').apply(np.argmin)
peak_days
```

```python
peak_days.to_csv('item_region_peak_days.csv')
```

```python
grouped_day_region = grouped_day[['d']].copy()
grouped_day_region['NE'] = grouped_day['HM-NE'] + grouped_day['CF-NE'] + grouped_day['FF-NE']
grouped_day_region['SW'] = grouped_day['HM-SW'] + grouped_day['CF-SW'] + grouped_day['FF-SW']
grouped_day_region['NW'] = grouped_day['HM-NW'] + grouped_day['CF-NW'] + grouped_day['FF-NW']
grouped_day_region['SE'] = grouped_day['HM-SE'] + grouped_day['CF-SE'] + grouped_day['FF-SE']
grouped_day_region['C'] = grouped_day['HM-C'] + grouped_day['CF-C'] + grouped_day['FF-C']
```

```python
grouped_day_region
```

```python
grouped_day_region.to_csv('grouped_day_region.csv')
```

```python
peak_days_reg = pd.DataFrame(grouped_day_region.set_index('d').apply(np.argmax),
                         columns=['peak_days'])
peak_days_reg['low_days'] = grouped_day_region.set_index('d').apply(np.argmin)
peak_days_reg
```

```python
peak_days_reg.to_csv('peak_days_reg.csv')
```

```python
grouped_day_item = grouped_day[['d']].copy()
grouped_day_item['HM'] = grouped_day['HM-NE'] + grouped_day['HM-SW'] +\
grouped_day['HM-NW']+ grouped_day['HM-SE'] + grouped_day['HM-C']
grouped_day_item['CF'] = grouped_day['CF-NE'] + grouped_day['CF-SW'] +\
grouped_day['CF-NW']+ grouped_day['CF-SE'] + grouped_day['CF-C']
grouped_day_item['FF'] = grouped_day['FF-NE'] + grouped_day['FF-SW'] +\
grouped_day['FF-NW']+ grouped_day['FF-SE'] + grouped_day['FF-C']
grouped_day_item
```

```python
grouped_day_item.to_csv('grouped_day_item.csv')
```

```python
grouped_day_total = grouped_day[['d']].copy()
grouped_day_total['Total'] = (grouped_day_item['HM'] + grouped_day_item['CF'] + grouped_day_item['FF'])/4
grouped_day_total
```

```python
grouped_day_total.to_csv('grouped_day_total.csv')
```

```python
total_sales_by_reg_item = pd.DataFrame(monthly.drop(
    columns=['Month, Year']).apply(sum))
total_sales_by_reg_item                                       
```

```python
total_sales_by_reg_item.to_csv('total_sales_by_reg_item.csv')
```

```python
region_stacked_bar = pd.DataFrame(
    total_sales_by_reg_item.values.reshape(-1, 3),
    columns=['Hamburger', 'Chicken Fillet', 'Fish Fillet'],
    index=['North East', 'South West', 'North West', 'South East', 'Central']
)
region_stacked_bar
```

```python
region_stacked_bar.to_csv('region_stacked_bar.csv')
```

```python
total_sales_by_date= pd.DataFrame(monthly.drop(
    columns=['Month, Year']).sum(axis=1))
total_sales_by_date['Date'] = monthly['Month, Year']
total_sales_by_date = total_sales_by_date.set_index('Date')
total_sales_by_date.head()
```

```python
total_sales_by_date.to_csv('total_sales_by_date.csv')
```

```python
total_sales_by_date_item = monthly[['Month, Year']].copy()
total_sales_by_date_item['HM'] = monthly['HM-NE'] + monthly['HM-SW'] +\
monthly['HM-NW']+ monthly['HM-SE'] + monthly['HM-C']
total_sales_by_date_item['CF'] = monthly['CF-NE'] + monthly['CF-SW'] +\
monthly['CF-NW']+ monthly['CF-SE'] + monthly['CF-C']
total_sales_by_date_item['FF'] = monthly['FF-NE'] + monthly['FF-SW'] +\
monthly['FF-NW']+ monthly['FF-SE'] + monthly['FF-C']
total_sales_by_date_item = total_sales_by_date_item.set_index('Month, Year')
total_sales_by_date_item.head()
```

```python
total_sales_by_date_item.to_csv('total_sales_by_date_item.csv')
```
